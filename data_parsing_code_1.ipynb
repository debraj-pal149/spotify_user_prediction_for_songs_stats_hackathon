```python
import shlex

test_df=pd.DataFrame(columns=[i.replace('"','') for i in (header)[0].split(" ")])
for i in rows:
    joined_row="".join(i)
    output = shlex.split(joined_row)
    temp_dict={ key : value  for key, value in zip([i.replace('"','') for i in (header)[0].split(" ")], (output[1:])) }
    test_df=test_df.append(temp_dict,ignore_index=True)
```
